```python
students_scores = [
    ("Alice", 85),
    ("Bob", 78),
    ("Charlie", 92),
    ("David", 88),
    ("Eva", 76),
    ("Frank", 90),
    ("Grace", 81)
]
```

```python
highest_score = students_scores[0]
for student in students_scores:
    if student[1] > highest_score[1]:
        highest_score = student
print(f"Highest score in the class: {highest_score[1]} (by {highest_score[0]})")
```

```python
total_score = 0
for student in students_scores:
    total_score += student[1]
average_score = total_score / len(students_scores)
print(f"Average score of the class: {average_score:.2f}")
```

```python
threshold = 80
students_above_threshold = []
for student in students_scores:
    if student[1] > threshold:
        students_above_threshold.append(student[0])
```

```python
student_names = [student[0] for student in students_scores]
student_names.sort()
print(f"Student names in alphabetical order: {', '.join(student_names)}")
```
